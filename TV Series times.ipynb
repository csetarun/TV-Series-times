```python
import pandas as pd # importing required packages
import numpy as np
```

```python
train_1 = pd.read_csv('train_1.csv') #importing csv files
train_2 = pd.read_csv('train_2.csv') 
train_1
```

```python
train_2 = train_2.rename(columns={'time_start':'watch_time_start','time_end':'watch_time_end'}) #Renaming columns
train_2
```

```python
new2_df = pd.concat([train_1,train_2],axis=1) #Joining two dataframes
new2_df
```

```python
_, i = np.unique(new2_df.columns, return_index=True) #Removing duplicate columns
new2_df = new2_df.iloc[:, i]
new2_df
```

```python
new2_df['Length'] = ((pd.to_datetime(new2_df['watch_time_end'])-pd.to_datetime(new2_df['watch_time_start']))/np.timedelta64(1,'s')).astype(int) #Calculating program time span
```

```python
new2_df['spareTime'] = new2_df.assest_id==new2_df.assest_id.shift(1) pd.to_datetime(new2_df.time_end.shift(1)) - pd.to_datetime(new2_df.watch_time_start)
```

```python
new2_df['spareTime']
```

```python
new2_df['Overlap'] = pd.to_datetime(new2_df[['watch_time_end','time_end']].min(axis=1))-pd.to_datetime(new2_df[['watch_time_start','time_start']].max(axis=1)) #Calculating overlap of watch time and program time
new2_df['spareTime'] = []
new2_df.apply(lambda r: (r['Overlap'] + r["SpareTime"]) if r[""] > 0 else 0, axis=1)
new2_df['Overlap'] = Overlap + spareTime
```

```python
new2_df['Overlap'] = (new2_df['Overlap']/np.timedelta64(1, 's')).astype(int) #Converting to seconds
```

```python
new2_df['Rate'] = new2_df['Overlap']/new2_df['Length'] #Calculating Rate
new2_df = new2_df.drop(['genre','time_start','time_end','watch_time_start','watch_time_end','Index'],axis=1) #Removing columns 
new2_df.to_csv('result.csv') #Exporting results
```
